```python
# Criando um gerador de números aleatórios com semente fixa

rng = np.random.default_rng(seed=1701) # seed para reprodutibilidade

x1 = rng.integers(10, size=6) # 6 inteiros aleatórios entre 0 e 9, array unidimensional
x2 = rng.integers(10, size=(3, 4)) # array bidimensional 3x4
x3 = rng.integers(10, size=(3, 4, 5)) # array tridimensional 3x4x5
```

```python
# Exibindo as propriedades do array x3

print('x3 ndim:', x3.ndim)
print('x3 shape:', x3.shape)
print('x3 size:', x3.size)
print('x3 dtype:', x3.dtype)
```
